```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'gx')
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'r--') # 'r--' 代表紅色的標線
plt.axis([0,5,0,1])   #圖表大小範圍，寬度由0到5，高度由0到1
plt.ylabel('Y')
plt.xlabel('X')
plt.legend(('price','passenger'),loc='upper right')
plt.show()
```
